### Q11. mean value of each column against each weather condition

```python
data.head()
```

```python
data.info()
```

```python
# data.groupby('weather_condition').mean()
```

```python
# Exclude non-numeric columns from the grouping
numeric_columns = data.select_dtypes(include=['number']).columns

# Group the numeric data by the 'Weather' column and calculate the mean for each group
grouped_data = data.groupby('weather_condition')[numeric_columns].mean()

# Display the result
print(grouped_data)
```
